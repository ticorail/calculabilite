```python
class PDA:
    def __init__(self, etats, alphabet_entree, alphabet_pile,
                 transitions, etat_initial, symbole_pile_initial,
                 etats_finaux):
        
        self.etats = etats
        self.alphabet_entree = alphabet_entree
        self.alphabet_pile = alphabet_pile
        self.transitions = transitions
        self.etat_initial = etat_initial
        self.symbole_pile_initial = symbole_pile_initial
        self.etats_finaux = etats_finaux

    def simuler(self, chaine):
        # état = (position, état courant, pile)
        stack_init = [self.symbole_pile_initial]
        pile_etats = [(0, self.etat_initial, stack_init)]

        while pile_etats:
            pos, etat, pile = pile_etats.pop()

            # condition d'acceptation
            if pos == len(chaine) and etat in self.etats_finaux and pile == [self.symbole_pile_initial]:
                return True

            symbole = chaine[pos] if pos < len(chaine) else None
            sommet = pile[-1] if pile else None

            # explorer transitions
            for (e, entree, top, new_e, remplacement) in self.transitions:
                if e == etat and top == sommet:

                    # transition avec symbole
                    if entree == symbole:
                        new_pile = pile[:-1]  # pop
                        if remplacement != "":
                         new_pile += remplacement.split()[::-1]  # inverse pour que le premier symbole soit en haut

                        pile_etats.append((pos + 1, new_e, new_pile))

                    # epsilon transition
                    elif entree is None:
                        new_pile = pile.copy()
                        pile_etats.append((pos, new_e, new_pile))

        return False
```

```python
transitions_parentheses = [
    ("q", "(", "Z", "q", "(Z"),
    ("q", "(", "(", "q", "(("),
    ("q", ")", "(", "q", ""),  
]

pda_parentheses = PDA(
    etats={"q"},
    alphabet_entree={"(", ")"},
    alphabet_pile={"Z", "("},
    transitions=transitions_parentheses,
    etat_initial="q",
    symbole_pile_initial="Z",
    etats_finaux={"q"}
)
```

```python
print(pda_parentheses.simuler("(()())"))  
print(pda_parentheses.simuler("((())"))   
print(pda_parentheses.simuler(")()("))    
print(pda_parentheses.simuler(""))        
```

```python
transitions_a_b = [
    # empile A pour chaque a
    ("q", "a", "Z", "q", "A Z"),
    ("q", "a", "A", "q", "A A"),
    # dépile A pour chaque b
    ("q", "b", "A", "q", ""),
    
]


pda_a_b = PDA(
    etats={"q"},
    alphabet_entree={"a", "b"},
    alphabet_pile={"Z", "A"},
    transitions=transitions_a_b,
    etat_initial="q",
    symbole_pile_initial="Z",
    etats_finaux={"q"}
)
```

```python
print(pda_a_b.simuler("aabb"))    
print(pda_a_b.simuler("aaabbb"))  
print(pda_a_b.simuler("ab"))      
print(pda_a_b.simuler("aab"))     
print(pda_a_b.simuler(""))        
```
